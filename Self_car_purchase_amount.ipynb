```python
X=car_df.drop(['Customer Name','Customer e-mail','Country','Car Purchase Amount'], axis = 1)  #axis=1 for columns and 0 for rows
```

```python
# data should be scaled/ normalized  
from sklearn.preprocessing import MinMaxScaler

scaler=MinMaxScaler()
X_scaled = scaler.fit_transform(X) 
```

```python
# y_scaled= scaler.fit_transform(y) for this to work the data should be of the shape (500,1) however currently it is (500,) 

#Reshape your data either using array.reshape(-1, 1) if your data has a single feature or array.reshape(1, -1) if it contains a single sample.
```

```python
y=y.values.reshape(-1,1)
```

```python
y.shape
```

```python
y_scaled= scaler.fit_transform(y)
```

```python
y_scaled
```

```python
#training the network
```

```python
#the data should be divided into training and testing
#the testing dataset should never be seen by the model
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y_scaled, test_size = 0.25) #to devide the data into 70-30 ratio of train and test
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_test.shape
```

```python
# build the model in a sequential form i.e layer by layer
#first the input layer then hidden and then output layer 
```

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense #for fully connected neural net
```

```python
model= Sequential()
model.add(Dense(25,input_dim = 5, activation = 'relu')) #input dim to be specified in the first layer
model.add(Dense(25, activation = 'relu'))
model.add(Dense(1, activation = 'linear'))#we need to predict the value which is leniar

```

```python
model.summary()
```

```python
# dense layer 1: 5 inputs fully connected to 25 neurons= 125 parameters and 25 parameters for the bias
# dense layer 2: 25 input neurons connected to 25 hidden layer and bias
# output layer: 25 hidden layer to 1 output neuron
```

```python
model.compile(optimizer= 'adam', loss = 'mean_squared_error')
```

```python
#epochs_hist = model.fit(X_train,y_train, epochs= 100, batch_size= 10, verbose = 1, validation_split = 0.2) # 25 sample per batch
```

```python
epochs_hist = model.fit(X_train,y_train, epochs= 20, batch_size= 25, verbose = 1, validation_split = 0.2) # 25 sample per batch
```

```python
# evaluate the model and visualize
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
#adding title

plt.title('Model Loss')
plt.ylabel('Training and validation loss')
plt.xlabel('epochs no')
plt.legend(['training loss','validation loss'])
```

```python
#by reducing the number of neurons the model struggles to gain intelligence
```

```python
# Gender, Age, Annual Salary, Credit Card Debt
#a customer has the following data and we need to predict from this data
X_test_sample = np.array([[1,20,50000, 10000, 600000]])

y_predict=model.predict(X_test_sample)
y_prdict_actual= model.predict(X_test)
```
